### <center> Introdução e objetivos </center>

<p>
O  projeto, vai se basear no modelo de descrição matemática da propagação de uma doença em uma população, o modelo $SIR$, desenvolvida por Kermack e McKendrick (1927). Tal modelo, divide a população, com uma quantidade fixa de N indivíduos, em 3 grupos: 
 </p>
 </br>
<ul>
<li> $S(t) =$ Pessoas saúdaveis  e suscetíveis a serem infectadas</li> 
<li> $I(t) =$ Pessoas Infectadas </li>
<li> $R(t) =$ Pessoas Recuperadas e Imunes </li>
</ul>


<p>
Cada um dos grupos descreve a mudança na população durante o periodo de tempo ultilizando termos de dois parâmetros,  $β\:e\:γ$. β descreve a taxa efetiva de contato da doença, ou seja, um indivíduo infectado entra em contato com βN outros individuos por uma unidade de tempo (na qual a fração sucetível a contrair a doença é dada por $\:S /\:N$). $γ\:$ é a taxa média de recuperação, em outras palavras, 1/ γ é o período médio de tempo durante o qual um indíviduo infectado pode tansmitir. 
</p>

\begin{align}
\frac{{\rm d}S}{{\rm d}t} & = -\beta S I\\
\frac{{\rm d}I}{{\rm d}t} & = \beta S I - \gamma I\\
\frac{{\rm d}R}{{\rm d}t} & = \gamma I\\
\end{align}

### Objetivo 
<p>Temos como objetivo, ultilizar o modelo para calcular e visualizar o comportamento da pandemia, durante o resto do ano de 2021. Além disso, queremos prever como será o seu comportamento nos periodos de festas e feriados, como carnaval, natal dentre outros.</p>
<p>Contudo, como o modelo $S.I.R$ &nbsp;não leva em conta a taxa de mortalidade e a taxa de pessoas assintomáticas, por conta disso, ultilziaremos outro modelo, que se baseia no metodo $S.I.R.A.S.D$ .</p>

## <center> <i>SEÇÃO $III$</i> </center>
</br>

### <center> Métodos </center>

<p>O método utilizado, como já mencionado, será uma variação do modelor SIR, o $S.I.R.A.S.D$ (Susceptible-Infected-Recovered for Asymptomatic-Symptomatic and Dead),  onde levaremos em conta os suscetíveis, infectados, recuperados sintomáticos e assintomáticos, e mortos.
</p>
</br>

<ol>
<li>Equação:</li>
$$\frac{dS}{dt} = -(\beta_AI_A + \beta_SI_S)*\frac{S}{N}$$
<ul>
<li>
Na equação temos a variação das pessoas suscetíveis em relação ao tempo, para isso levamos em conta que parte deles se tornarão infectados assintomáticos$(IA)$ e outra parte será sintomática. Levamos em conta que essa variação irá depender no número total de sucetíveis , da quantidade de categoria de infectados$( I_A\: e\: I_B)$ , e também de constantes de contato entre infectados e sucetíveis $(\beta_A \: e \: \beta_B)$.
</li>
</ul>

</br>

<li>Equação</li>
$$\frac{dI_A}{dt} = (1 - p)*(\beta_AI_A + \beta_SI_S)*\frac{S}{N}-(\gamma_A)I_A$$
<ul>
<li>
Primeiramente definiremos $p$ como a proporção de indivíduos infectados que irão desenvolver sintomas, dessa forma $(1-p)$ é a quantidade que será assintomática. a primeira parcela equação diz respeito à parcela dos infectados totais que irão se tornar assintomáticos  que irão somar à variação dos casos assintomáticos, já a segunda parcela diz respeito ao número de recuperados dos assintomáticos, que deixaram de ser infectados e por isso são subtraídos.
</li>
</ul>


</br>

<li>Equação</li>
$$\frac {dI_s}{dt} = p*(\beta_AI_A + \beta_SI_S)*\frac{S}{N}-\gamma_SI_S - \frac{\rho}{1-\rho}*\gamma_SI_S =\\ p*(\beta_AI_A + \beta_SI_S)*\frac{S}{N}-\frac{\gamma_SI_S}{1-\rho}$$
<ul>
<li>
Analogamente à equação 2, essa equação corresponde à variação de infectados sintomáticos em relação ao tempo. Agora teremos uma  parcela a mais, pois iremos considerar que alguns dos sintomáticos irão morrer, onde $ρ$ significa a probabilidade de uma pessoa infectada morrer antes de ser recuperada.
</li>
</ul>

</br>

<li>Equação</li>
$$\frac{dR_A}{dt} = \gamma_AI_A$$
<ul>
<li>
A equação 4 trata-se da variação do número de pessoas recuperadas que eram infectadas assintomáticas, para isso consideramos que essa variação será diretamente proporcional ao número de infectados assintomáticos, por meio de uma constante $γ_A$ 
</li>
</ul>

</br>

<li>Equação</li>
$$\frac{dR_S}{dt} = \gamma_SI_S$$
<ul>
<li>
A equação 5, trata-se da variação do número de recuperadas que eram infectadas sintomáticas.Para isso consideramos que essa variação será diretamente proporcional ao número de infectados sintomáticos, por meio de uma constante $γ_S$ .</li>
</ul>

</br>

<li>Equação</li>
$$\frac{dD}{dt} = \frac {\rho}{1-\rho}*\gamma_SI_S$$
<ul>
<li>
A equação 6, trata-se da variação do número de mortes .Para isso consideramos que essa variação será diretamente proporcional ao número de infectados sintomáticos. é usado como coeficiente $$\frac {\rho}{1-\rho}$$, pois a morte depende da quantidade de pessoas que morrem por unidade de tempo em relação ao numero de indivıduos infectados no mesmo perıodo (p número de mortos e (1-p) infectados que não morrem)
</li>
</ul>

</ol>

```python
#import das bib
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#Constantes(os parâmetros)
BetaA = 0.4417; BetaS = 0.4417
GamaA = 0.1260; GamaS = 0.1508
ro = 0.0347
N = 1000 # total population
p = 0.3210

#Condições iniciais
RAo = 2.4209; RSo = 1.1807
#Ro = 3.6017
ISo = 1; IAo = 0
So = N - IAo - ISo
Do =  0 
t = np.linspace(0,200,10001)
```

```python
def deriv(funcao, t, N, BetaA, BetaS, GamaA, GamaS, ro, p):
  S, Ia, Is, Ra, Rs, D = funcao
  dSdt = -(BetaA*Ia + BetaS*Is) * (S/N)
  dIadt = (1-p)*(BetaA*Ia + BetaS*Is) * (S/N) - GamaA*Ia
  dIsdt = p*(BetaA*Ia + BetaS*Is) * (S/N) - ((GamaS*Is)/(1-p))
  dRadt = GamaA * Ia
  dRsdt = GamaS * Is
  dDdt = (p/(1-p)) * GamaS * Is
  return [dSdt, dIadt, dIsdt, dRadt, dRsdt, dDdt]

#Vetor de condições iniciais
y0 = [So, IAo, ISo, RAo, RSo, Do]
# Integrando as equações SIRASD em função de t.
solucao = odeint(deriv, y0, t, args=(N, BetaA, BetaS, GamaA, GamaS, ro, p))
    
S, Ia, Is, Ra, Rs, D = solucao.T
```
